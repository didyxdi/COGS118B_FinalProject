```python
import pandas as pd
import numpy as np
from numpy import nan as Nan
```

```python
IP2010 = pd.read_excel("datasets/before/2010IP.xlsx").fillna(0) # Inpatient Discharges Data in 2010

IP2010.head()
# Note: To preview the data in the folder, you should change ".xlsx" back to ".csv"
```

```python
IP2010_df = pd.DataFrame()
IP2010_df[["1q10"]] = IP2010.loc[:,["Hospital county", "1q10 Inpatient Discharges"]].groupby("Hospital county").sum()
IP2010_df[["2q10"]] = IP2010.loc[:,["Hospital county", "2q10 Inpatient Discharges"]].groupby("Hospital county").sum()
IP2010_df[["3q10"]] = IP2010.loc[:,["Hospital county", "3q10 Inpatient Discharges"]].groupby("Hospital county").sum()
IP2010_df[["4q10"]] = IP2010.loc[:,["Hospital county", "4q10 Inpatient Discharges"]].groupby("Hospital county").sum()
IP2010_df.index.names = ["county"]
```

```python
IP2010_df # Total inpatient discharges each quarter in 2010 grouped by county
```

```python
social2010_df = pd.read_csv("datasets/before/2010social.csv").fillna(0) # Social Factors in 2010
social2010_df.columns = social2010_df.iloc[0] # Set the top row as headers
social2010_df = social2010_df[1:]

social2010_df.head()
```

```python
def convert_county_name(input_county):
    return input_county[:input_county.find(" County")]

def convert_id(input_id):
    return input_id[9:]

social2010_df = social2010_df.rename(columns={"Geographic Area Name":"county"})
social2010_df["county"] = social2010_df["county"].apply(convert_county_name)
social2010_df["id"] = social2010_df["id"].apply(convert_id)

social2010_df = social2010_df.set_index("county");
```
